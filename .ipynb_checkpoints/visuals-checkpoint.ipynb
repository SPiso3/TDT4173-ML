# DATA VISUALIZATION

```python
import pandas as pd 
import xgboost as xgb
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
pd.set_option('display.max_rows', None) 
```

```python
train = pd.read_csv('ais_train.csv', sep='|')
land = gpd.read_file("ne_10m_land.zip")
```

```python
land.plot()
```

```python
land.total_bounds
```

```python
geometry = [Point(xy) for xy in zip(train['longitude'], train['latitude'])]
geo_df = gpd.GeoDataFrame(train, geometry=geometry, crs="EPSG:4326")
geo_df.set_crs(land.crs, inplace=True)
```

```python
geo_df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

land.plot(ax=ax)

# Plot the points with a specified size and color
geo_df.plot(ax=ax, color='red', markersize=1)
```

```python
geo_df['on_land'] = geo_df.geometry.apply(lambda point: land.contains(point).any())
```
